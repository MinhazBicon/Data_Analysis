```python
#change AVD_DATE tao date time
modified_data["ADV_DATE"] = pd.to_datetime(modified_data["ADV_DATE"])
```

```python
modified_data.info()
```

```python
year = modified_data["ADV_DATE"].apply(lambda b:b.year)
```

```python
month = modified_data["ADV_DATE"].apply(lambda b:b.month)
```

```python
modified_data["Year"]=year
modified_data["Month"]=month
```

```python
modified_data.head()
```

```python
# convert month name
dmap = {1:"JAN",2:"FEB",3:"MAR",4:"APR",5:"MAY",6:"JUN",7:"JUL",8:"AUG",9:"SEP",10:"OCT",11:"NOV",12:"DEC"}
```

```python
modified_data["Month"] = modified_data.Month.map(dmap)
```

```python
modified_data.Month.value_counts()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Month',data=modified_data,hue="REGION")
```

```python
modified_data.groupby(["Month","REGION"]).count()
```

```python
from shapely.geometry import Point
import geopandas as geo
```

```python
#creating a funtion, geographical plot
def geo_plot (data):
    geometry= [Point(xy) for xy in zip(data["LONG_"],data["LAT"])] 
    gdf  = geo.GeoDataFrame(data,geometry=geometry)
    #world map
    world=geo.read_file(geo.datasets.get_path('naturalearth_lowres'))
    #ploting
    gdf.plot(ax=world.plot(figsize=(20,20)),marker='o',color='red',markersize=10)

```

```python
geo_plot(modified_data)
```
