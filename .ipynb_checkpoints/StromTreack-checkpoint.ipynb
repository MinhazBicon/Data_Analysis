```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data import
storm = pd.read_csv("Strom track.csv")
```

```python
# showing data
storm.head(12)
```

```python
# cloolect information
storm.info()
```

```python
storm.isnull().sum()
```

```python
# data cleaning
modified_data = storm.dropna()
```

```python
storm.head()
```

```python
modified_data.isnull().sum()
```

```python
modified_data["REGION"].value_counts()
```

```python
import seaborn as sns

%matplotlib inline
```

```python
sns.countplot(x='REGION',data=modified_data)
```

```python
#change AVD_DATE tao date time
modified_data["ADV_DATE"] = pd.to_datetime(modified_data["ADV_DATE"])
```

```python
modified_data.info()
```

```python
year = modified_data["ADV_DATE"].apply(lambda b:b.year)
```

```python
month = modified_data["ADV_DATE"].apply(lambda b:b.month)
```

```python
modified_data["Year"]=year
modified_data["Month"]=month
```

```python
modified_data.head()
```

```python
# convert month name
dmap = {1:"JAN",2:"FEB",3:"MAR",4:"APR",5:"MAY",6:"JUN",7:"JUL",8:"AUG",9:"SEP",10:"OCT",11:"NOV",12:"DEC"}
```

```python
modified_data["Month"] = modified_data.Month.map(dmap)
```

```python
modified_data.Month.value_counts()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='Month',data=modified_data,hue="REGION")
```

```python
modified_data.groupby(["Month","REGION"]).count()
```
